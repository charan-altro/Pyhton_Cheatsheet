## Escape sequence:

- if we want to print double or single quotes inside double and single quote respectively then use the escape sequences  


```python
print( "hello \"world\" world")                         
```

## Escape Sequence---->Meaning--->notes
                                    


### \\ < newline > --->	Backslash and newline ignored

```python
print("a\
b")
```

### \\\\--->	Backslash (\\)

```python
print(" Name\\Backslash\\") 
print ("\\")
```

### \\'	Single quote (') and \\"	Double quote (")

```python
print( "hello \"world\" \'world\'") 
```

### ASCII Horizontal Tab (TAB)

```python
#\t	ASCII Horizontal Tab (TAB)	
print(" Hello\tWorld")                                     
```

### ASCII Backspace (BS)

```python
#\b	ASCII Backspace (BS)	
print("Hell \bo world")
print ("ab" + "\b" + "c")
```

### ASCII Bell (BEL)

```python
#\a	ASCII Bell (BEL)	
print("Hello world\ax\ay\az\a")
```

### ASCII Formfeed (FF)

```python
#\f	ASCII Formfeed (FF)	
print('Hello \fworld')
```

### ASCII Linefeed (LF)

```python
#\n	ASCII Linefeed (LF)	
print(" line A \n line B")  
```

```python
#\r	ASCII Carriage Return (CR)	
''''ASCII carriage return (CR). 
Moves all characters after ( CR ) the the beginning of the 
line while overriding same number of characters moved.'''
print ("123456\rXX_XX")
```

```python

#\v	ASCII Vertical Tab (VT)	
print('Hello \vworld')

#\ooo	Character with octal value ooo	(1,3)
print ("\043xxx")

#\xhh	Character with hex value hh	(2,3)
print ("\x23yyy")


'''Escape Sequence--------->Meaning--->Notes

\N{name}------------->Character named name in the Unicode database-->(4)

\uxxxx-------------->Character with 16-bit hex value xxxx--->(5)

\Uxxxxxxxx-------------->Character with 32-bit hex value xxxxxxxx--->(6)'''
```
