# What is REPL? 

### REPL is the language shell, the Python Interactive Shell. 
### The REPL acronym is short for Read, Eval, Print and Loop
### The process is:

#### Read: take user input.

#### Eval: evaluate the input.

#### Print: shows the output to the user.

#### Loop: repeat.

### *Read function* accepts an expression from the user and parses it into a data structure in memory

```python
x = 5
```

### The eval function takes this internal data structure and evaluates it

```python
y = x * 5
```

### The print function takes the result yielded by eval, and prints it out to the user.

```python
print("x is {}\ny is {}".format(x,y))
```

#### REP example

```python
print( _ *1) # _ (underscore will return last max excuted value)
print( _ *2)
```

## %S 
### You can use the %s to format strings into your print statements.

```python
s = 'STRING'
print ('Place another string with a mod and s: %s' %(s))

print('%s %s' % ('one', 'two'))
```

## %d
### You can use the %d to format integers into your print statements.

```python
print('%d %d' % (5, 2))
```

## %1.nf  

### (%n1.n2f )--- n1 is the total minimum number of digits the string should contain and Floating Point Numbers at n2 decimal places


```python
print ('Floating point numbers: %1.2f' %(13.144))
print ('Floating point numbers: %1.0f' %(13.144)) #Truncating long strings
print ('Floating point numbers: %1.5f' %(13.144))
print ('Floating point numbers: %10.2f' %(13.144))  # 10 places alligning will happen (Padding and aligning strings)
print ('Floating point numbers: %25.2f' %(13.144)) # 25 places alligning will happen (Padding and aligning strings)
```

# Conversion Format methods-->%s and %r --> str() and repr()

## .format()
### Using the string .format() method

```python
print('String here {var1} then also {var2}'.format(var1='something1',var2='something2'))
print ('This is a string with an {p}'.format(p='insert in variable'))
# Multiple times:
print ('One: {p}, Two: {p}, Three: {p}'.format(p='Hi!'))
# Several Objects:
print ('Object 1: {a}, Object 2: {b}, Object 3: {c}'.format(a=1,b='two',c=12.3))
print('{} {}'.format(1, 2))
#.format allows for re-arranging the order of display without changing the arguments.
print('{1} {0}'.format('one', 'two'))
# Padding and aligning strings using .formot method
print('{:>30}'.format('test for Align right'))
print('{:10}'.format('test for Align left'))
print('{:_<50}'.format('test for Align left')) # only .format we can achive this, not able old methods
print('{:_^100}'.format('test center align values'))
print('{:_^6}'.format('zip'))# uneven split of the padding characters the extra character will be placed on the right side
print('{:.5}'.format('xylophone')) #Truncating long strings
print('{:_^10.5}'.format('xylophone'))#Combining truncating and padding
```
